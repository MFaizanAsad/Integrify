### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
# This 
df = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user", delimiter = '|')
df.head()
df.set_index('user_id', True)
```

### Step 4. Discover what is the mean age per occupation

```python
df.groupby('occupation')['age'].mean().to_frame()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
gender_ratio = (df.groupby('occupation')['gender'].value_counts('M')).sort_values(ascending = False).xs("M",level =1)
print(gender_ratio)


```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
df.groupby('occupation')['age'].agg(['min', 'max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python

# df['occupation'].value_counts()['M']
combination_mean_age=df.groupby(['occupation','gender'])['age'].mean().to_frame()
combination_mean_age
```

### Step 8.  For each occupation present the percentage of women and men

```python
g= df.groupby('occupation')['gender'].value_counts('M')
gr= g.to_frame()
(gr['gender']*100).to_frame()

```
