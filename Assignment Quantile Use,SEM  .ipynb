2.Explain in your own words what is standard error of the mean?

Answer:
***Standard Error of the Mean (SEM)***
The standard error of statistic (usually an estimate of a parameter) is the standard deviation of its sampling distribution. IF the parameter of the statistic is the mean, it is called standard error of the mean (SEM).

The standard error of the mean is basically the standard deviation of the sampling distribution of the mean.
