```python
#Task 5 Decorator1
'''Write a function called double. The function should decorate a function and return two copies of it's return value in a list.
'''
def double(fun):
    def inner():
        #print("\n")
        fun()
        fun()
        #print("\n")
    return inner
@double 
def fun():
    print("DIITO")
    
fun()
```
